```python
import torch
import timeit
from torch import nn, optim
from torch.autograd.variable import Variable
from torchvision import transforms, datasets
from utils import Logger
```

```python
def load_data():
    data_trans = transforms.Compose([transforms.ToTensor(), transforms.Normalize((.5,),(.5,))])
    out_dir = './data'
    data = datasets.MNIST(root = out_dir, transform = data_trans, train= True, download=True)
    
    return data
```

```python
# Load Data
real_image_set = load_data()
device = "cuda"
```

```python
# Discriminator Network
class DiscriminatorNet(nn.Module):
    
    def __init__(self):
        super(DiscriminatorNet, self).__init__()
        n_features = 28*28
        
        self.layer0 = nn.Sequential(nn.Linear(n_features, 1024),
                                   nn.LeakyReLU(0.2), 
                                   nn.Dropout(0.3))
        self.layer1 = nn.Sequential(nn.Linear(1024, 512),
                                   nn.LeakyReLU(0.2), 
                                   nn.Dropout(0.3))
        self.layer2 = nn.Sequential(nn.Linear(512, 256),
                                   nn.LeakyReLU(0.2), 
                                   nn.Dropout(0.3))
        self.output = nn.Sequential(nn.Linear(256, 1),
                                   nn.Sigmoid())
        
    def forward(self, x):
        x = self.layer0(x)
        x = self.layer1(x)
        x = self.layer2(x)
        x = self.output(x)
        
        return x
        
```

```python
# Generator Network
class GeneratorNet(nn.Module):
    
    def __init__(self):
        super(GeneratorNet, self).__init__()
        n_features = 100
        n_out = 28*28
        self.layer0 = nn.Sequential(nn.Linear(n_features, 256),
                                   nn.LeakyReLU(0.2))
        self.layer1 = nn.Sequential(nn.Linear(256, 512),
                                   nn.LeakyReLU(0.2))
        self.layer2 = nn.Sequential(nn.Linear(512, 1024),
                                   nn.LeakyReLU(0.2))
        self.output = nn.Sequential(nn.Linear(1024, n_out),
                                   nn.Tanh())
        
    def forward(self, x):
        x = self.layer0(x)
        x = self.layer1(x)
        x = self.layer2(x)
        x = self.output(x)
        
        return x
```

```python
# Create fake_images by sampling from noise
def noise(n_samples):
    '''
    Each noise sample is a latent data of 100 dimensions
    '''
    return Variable(torch.randn(n_samples, 100))
```

```python
# Create targets 1 for real images and 0 for fake images
def real_targets(n_samples):
    return Variable(torch.ones(n_samples, 1))#.long()

def fake_targets(n_samples):
    return Variable(torch.zeros(n_samples, 1))#.long()
```

```python
# Changing dimensions: Change img to vector and vector to img
def img2vec(images):
    return images.view(images.size(0), 784)

def vec2img(vectors):
    return vectors.view(vectors.size(0),1,28,28)
```

```python
discriminator = DiscriminatorNet().to(device)
loss = nn.BCELoss().to(device) #nn.CrossEntropyLoss()
```

```python
# Discriminator Trainer
def train_discriminator(optimizer, real_images, fake_images):
    
    n_samples = real_images.size(0)
    # Clear the gradients for the optimizer
    optimizer.zero_grad()
    
    # Pass the real_images and see what the discriminator predicts for it
    real_pred = discriminator(real_images)
    
    # Measure it against an expectation of predicted prob == 1
    error_vs_real = loss(real_pred, real_targets(n_samples).to(device))
    
    # Back prop the error
    error_vs_real.backward()
    
    # Pass the fake_images and see what the discriminator predicts for it
    fake_pred = discriminator(fake_images)
    
    # Measure it against the expectation of a zero value to be predicted
    error_vs_fake = loss(fake_pred, fake_targets(n_samples).to(device))
    
    # Back prop the error
    error_vs_fake.backward()
    
    # Push an update step based on the back prop-d gradients
    optimizer.step()
    
    return error_vs_real+error_vs_fake, real_pred, fake_pred
```

```python
# Generator Trainer
def train_generator(optimizer, fake_images):
    
    # Clear the gradients for the optimizer
    optimizer.zero_grad()

    # Pass the fake_images and see what the discriminator predicts for it
    fake_pred = discriminator(fake_images)
    
    # Measure it against the expectation of a one value to be predicted(cheating D)
    error_vs_fake = loss(fake_pred, real_targets(fake_images.size(0)).to(device))
    
    # Back prop the error
    error_vs_fake.backward()
    
    # Push an update step based on the back prop-d gradients
    optimizer.step()
    
    return error_vs_fake
```

```python
# Train
```

```python
# Train Discriminator and Generator together
def train():
    '''
    We train D and G one after the other in every iteration.
    First comes D: 
        - Real data comes directly from the data input
        - Fake data is synthesized and flow through G. But G part is detached from the grad tree
        - Real and Fake data are passed there by the perdiction of D against real and fake will only update the D net
    
    Next its G:
        - The fake data is synthesized and flow throught G. Here G is retained in the Grad_graph
        - Then the output from G is passed for trial through D
        - Based on its prediction error, the entire chain D and G together is updated        
    '''
    n_epochs = 100

    
    generator = GeneratorNet().to(device)

    dis_optimizer = optim.Adam(discriminator.parameters(), lr = 0.0002)
    gen_optimizer = optim.Adam(generator.parameters(), lr = 0.0002)
    
    # Create a data loader
    data_loader = torch.utils.data.DataLoader(real_image_set, batch_size = 100, shuffle = True)
    num_batches = len(data_loader)
    
    num_test_samples = 16
    test_noise = noise(num_test_samples).to(device)

    # Create logger instance
    logger = Logger(model_name='VGAN', data_name='MNIST')
    
    for itr in range(n_epochs):
        start = timeit.time.time()
        # load a sample from the data loader
        for batch_idx, (real_images, _) in enumerate(data_loader):
            n_samples = real_images.size(0)
            # Train D
            real_data = Variable(img2vec(real_images)).to(device)
            fake_data = generator(noise(n_samples).to(device)).detach()
            
            d_error, real_pred, fake_pred = train_discriminator(dis_optimizer, real_data, fake_data)
            
            # Train G
            fake_data = generator(noise(n_samples).to(device))
            
            g_error = train_generator(gen_optimizer, fake_data)
            
            # Generate and visualize output on a fixed seed of noise samples
            # log the values            
            logger.log(d_error, g_error, itr, batch_idx, num_batches)
            
            # Once for every bulk of iterations, visualize how good G is performing
        
        end = timeit.time.time()
        if (itr) % 10 == 0: 
            
            test_images = vec2img(generator(test_noise))
            test_images = test_images.data
            logger.log_images(
                test_images.to('cpu'), num_test_samples, 
                itr, batch_idx, num_batches
            );
            # Display status Logs
            logger.display_status(
                itr, n_epochs, batch_idx, num_batches,
                d_error, g_error, real_pred, fake_pred
            )
            
            print("Time taken for this epoch: {0} secs".format(end-start))


            
    
```

```python
train()
```
